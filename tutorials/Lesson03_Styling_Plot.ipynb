# 3 Styling plot

In previous section, we have seen how to show data in various plots, but we just use the default setting of the plot style.

Sometimes, we may way to change the style of the plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
root_path="https://minio.lab.sspcloud.fr/pengfei/diffusion/data_viz"

input_file=f'{root_path}/winemag-data_first150k.csv'
reviews=pd.read_csv(input_file,index_col=0)
reviews.head(3)
```

## 3.1 Set up plot size

Below figure shows a bar plot of the vin score(points column) with default
settings.

```python
vin_points=reviews['points'].value_counts().sort_index()
vin_points.plot.bar()
plt.show()
```

We could say the plot is too small to see, we want to make it big
we can use **figsize(width, height)** parameter to define the size of plot


```python
vin_points.plot.bar(figsize=(12,6))
plt.show()
```

## 3.2 Change bar color

Now, we want to change the color of bars. We can use **color='color_value'**. The color_value is a string, the possible value could be (mediumvioletred, red, yellow, etc)

```python
vin_points.plot.bar(figsize=(12,6),color='red')
plt.show()
```

## 3.3 Change the text fontsize

The text label are very hard to read, we can change the fontsize with **fontsize** parameter.

```python
vin_points.plot.bar(figsize=(12,6),color='yellow',fontsize=16)
plt.show()
```

## 3.4 Add a title to the plot

We can add title by using the parameter **title**.

```python
vin_points.plot.bar(figsize=(12, 6), color='mediumvioletred', fontsize=16, title='Rankings Given by Wine Magazine')
plt.show()
```

## 3.5 Use matplotlib function

Pandas data visu is built based on matplotlib, so all the plot build with pandas can use also the function of matplotlib

The following example we will use matplotlib function to set title with a bigger font size

```python
my_plot=vin_points.plot.bar(figsize=(12, 6), color='mediumvioletred', fontsize=16)
my_plot.set_title("Rankings Given by Wine Magazine",fontsize=20)
plt.show()
```

## 3.6 Seaborn function

We can use another lib called seaborn which is also built based on matplotlib

```python
my_plot=vin_points.plot.bar(figsize=(12, 6), color='mediumvioletred', fontsize=16)

my_plot.set_title("Rankings Given by Wine Magazine",fontsize=20)

# sns.despine method to turn off the ugly black border.
sns.despine(bottom=False,left=True)

plt.show()
```
